```python
import os
import re
import rasterio
import numpy as np
from tqdm import tqdm

# Function to extract year from filenames
def extract_year(filename):
    match = re.search(r'-(\d{4})-', filename)  # Regex to match year in the format "-YYYY-"
    if match:
        return int(match.group(1))  # Return the year as an integer
    return None

# Function to generate the binary mask of undisturbed areas
def generate_undisturbed_mask_chunked(yearly_images_paths, output_path):
    with rasterio.open(yearly_images_paths[0]) as src:
        meta = src.meta
        height, width = src.height, src.width
        meta.update(dtype=rasterio.uint8, count=1)  # Update metadata for binary output
        block_size = 1024  # Define chunk size for processing (adjust as needed)

    # Create the output file
    with rasterio.open(output_path, 'w', **meta) as dst:
        # Add progress tracking with tqdm
        total_chunks = (height + block_size - 1) // block_size
        with tqdm(total=total_chunks, desc="Processing Chunks", unit="chunk") as pbar:
            for y_start in range(0, height, block_size):
                y_end = min(y_start + block_size, height)
                consistent_chunk = None
                for i, image_path in enumerate(yearly_images_paths):
                    with rasterio.open(image_path) as src:
                        # Read a chunk (window) of data
                        window = rasterio.windows.Window(0, y_start, width, y_end - y_start)
                        data = src.read(1, window=window)

                        if i == 0:
                            consistent_chunk = np.ones_like(data, dtype=bool)  # Initialize as all True
                            reference_chunk = data  # Use the first year's data as a reference
                        else:
                            consistent_chunk = np.logical_and(consistent_chunk, data == reference_chunk)

                # Convert consistent_chunk to uint8 (binary format)
                mask_chunk = consistent_chunk.astype(np.uint8)
                dst.write(mask_chunk, 1, window=window)  # Write chunk to the output file
                pbar.update(1)  # Update progress bar

    print(f"Undisturbed areas binary image saved to {output_path}")

# Define input and output paths
yearly_images_directory = r'C:\Users\llccl\OneDrive\Documents\To Do\Academics\2nd Year - 1st Semester\Master Thesis\2 - Thesis Proper\5 - Processing\1. Converted Copernicus LC Gridded'
export_directory = r'C:\Users\llccl\Downloads\Scripts\test'
output_binary_image = os.path.join(export_directory, 'undisturbed_areas_2006_2015.tif')

# Extract and sort yearly image paths
files = [f for f in os.listdir(yearly_images_directory) if f.endswith('.tif')]
files_with_years = [(f, extract_year(f)) for f in files if extract_year(f) is not None and 2006 <= extract_year(f) <= 2015]
files_with_years.sort(key=lambda x: x[1])  # Sort by year
yearly_images_paths = [os.path.join(yearly_images_directory, f[0]) for f in files_with_years]

# Verify if files are correctly sorted
print("Yearly image paths (sorted):")
for path in yearly_images_paths:
    print(path)

# Generate the binary mask
generate_undisturbed_mask_chunked(yearly_images_paths, output_binary_image)
```
